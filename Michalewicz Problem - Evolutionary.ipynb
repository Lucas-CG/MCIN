```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


def Func_obj1(X):
    z = []
    for cromo in X:
        sum_x = 0
        for index in range(len(cromo)):
            sum_x -= np.sin(cromo[index])*((np.sin((index+1)*(cromo[index]**2)/math.pi))**(2*10))
        z.append(sum_x)
    
    return z

def Product_list(list1, list2, list3, F):
    vec_list = []
    for index in range(len(list1)):
        vec_list.append(list1[index] + F*(list3[index] - list2[index]))
    
    return vec_list

def Check_factible(vector,mininum,maximum):
    test_fact = 0  #Verify if it is factible#

    for index in range(len(vector)):
        if mininum[index] > vector[index] or vector[index] > maximum[index]:
            test_fact += 1
    
    return test_fact
    
    
def Muta_vector(X_parameters, Func_values, F):
    mutated = []
    
    while len(mutated) < len(Func_values):
        
        for index in range(0,len(Func_values)):
            print("The current index is: ", index)
            
            while len(mutated) != index+1:

                r_n = random.sample(range(0, len(Func_values)), 3)

                print(r_n)

                while index in r_n:
                    r_n = random.sample(range(0, len(Func_values)), 3)
                    print(r_n)

                x_r1 = X_parameters[r_n[0]]
                x_r2 = X_parameters[r_n[1]]
                x_r3 = X_parameters[r_n[2]]

                print("The intermediary vectors are: ", x_r1, x_r2, x_r3)

                result = Product_list(x_r1, x_r2, x_r3, F)

                print("The resultant vectors is: ", result)

                
                test_fact = Check_factible(result,min_values,max_values)

                if test_fact == 0:
                    print("Mutated vector is factible")
                    mutated.append(result)

                else:
                    print("Mutated vector is not factible")
                    
                print("Mutated vector: ", result)
    
    mutated_f = Func_obj1(mutated)
    
    return mutated, mutated_f

def Crossed_vector(X_parameters, mutated_par,CR):
    trial_pop = []
    
    pass
    for index in range(len(X_parameters)):
        xi = X_parameters[index]
        vi = mutated_par[index]
        trial = []
        
        print("The curent cromossom index is: ", index)
        
        j_rand = random.sample(range(0, len(mutated_par)), 1)
        
        print("The selected j parameter is: ", j_rand)
        
        for index2 in range(len(xi)):
            cross_over_test = random.random()
            
            print("The current parameter in index is: ", index2)

            if cross_over_test < CR or index2 == j_rand:
                print("Crossed")
                trial.append(vi[index2])
            
            else:
                print("Not Crossed")
                trial.append(xi[index2])
        
        trial_pop.append(trial)
        
    print("The trial population is: ", trial_pop)
        
    trial_pop_func = Func_obj1(trial_pop)
    
    print("The trial population objective values are: ", trial_pop_func)
        
    return trial_pop, trial_pop_func

def Selection(X_parameters, X_func, mutated_par, mutated_func):
    selected = []
    
    for index in range(len(X_func)):
        if X_func[index] < mutated_func[index]:
            print("The vector selected is the one from the original population")
            selected.append(X_parameters[index])
        
        else:
            print("The vector selected is the one from the muta/cross population")
            selected.append(mutated_par[index])
    
    selected_apt = Func_obj1(selected)
    
    return selected, selected_apt
        


average_apt = []
maximum_apt = []
min_values = [0,0]
max_values = [math.pi,math.pi]


number_pop = int(input("Enter the size of the population (NP):  "))
number_gen = int(input("Enter the number of generations:  "))    
cross_rate = float(input("Enter the tax for crossover (60%-90%): "))/100
F_step = float(input("Enter the size of step F: "))
number_elit = int(input("Enter the number of elit parents in each generation: "))


def Genetic_Alg(initial_pop, generations, elitism, F, CR):
    
    first_pop_x = []

    for size in range(initial_pop):
        variables = []
        for index in range(len(min_values)):
            variables.append(random.uniform(min_values[index],max_values[index]))
        
        first_pop_x.append(variables)
        
    
    current_pop = first_pop_x
    
    for i in range(generations):
        
        print("\n Start of generation number: ", i, "\n")
                
        current_pop_y = Func_obj1(current_pop)
        
        print("The current population is: ", current_pop)
        print("The object function is: ", current_pop_y)
        
        mutated_vector, mutated_apt = Muta_vector(current_pop, current_pop_y, F)
        
        print("The mutated candidates are: ", mutated_vector)
        print("The mutated candidates aptitude are: ", mutated_apt)
        
        trial_pop, trial_pop_func = Crossed_vector(current_pop, mutated_vector,CR)
        
        print("The population after trial is: ", trial_pop)
        print("The aptitudes after trial are: ", trial_pop_func)
        
        selected_pop, selected_pop_func = Selection(current_pop, current_pop_y, trial_pop, trial_pop_func)
        
        print("The population after selection is: ", selected_pop)
        print("The aptitudes after selection are: ", selected_pop_func)
        
        maximum_apt.append(max(selected_pop_func))
        average_apt.append(np.mean(selected_pop_func))
        
        current_pop = selected_pop
        
        
        
Genetic_Alg(number_pop, number_gen,number_elit, F_step, cross_rate)

generation = np.arange(1,number_gen+1)

print(average_apt)
print(maximum_apt)

graph1 = plt.plot(generation, average_apt, label = "Average Aptitude")
grapt2 = plt.plot(generation, maximum_apt, label = "Maximum Aptitude")

plt.grid('ON',linestyle='--', alpha = 0.3)

plt.xlabel("Number of generations")
plt.ylabel("Fitness")

plt.legend()

plt.show()
```
